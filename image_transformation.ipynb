```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img = cv2.imread('/content/dog.jfif')
height, width = img.shape[:2]
res = cv2.resize(img,(2*width,5*height),interpolation = cv2.INTER_NEAREST)
plt.imshow(img)
```

```python
#Transformation
img = cv2.imread('/content/dog.jfif')
rows, cols, s = img.shape
M = np.float32([[1,0,200],[0,1,100]])
dst = cv2.warpAffine(img,M,(cols,rows))
plt.imshow(dst)
```

```python
#Rotation
M = cv2.getRotationMatrix2D((cols/2,rows/2),50,1)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.imshow(dst,'gray')
```

```python
#Morphological
img = cv2.imread('/content/dog.jfif')
kernel = np.ones((5,5),np.uint8)
plt.imshow(img,'gray')
```

```python
#Erotion
erosion = cv2.erode(img,kernel,iterations=1)
plt.imshow(erosion,'gray')
dilation = cv2.dilate(img,kernel,iterations=1)
plt.imshow(dilation,'gray')
```

```python
#Dilation
img = cv2.imread('/content/dog.jfif')[0:150,0:120]
plt.imshow(img)
```

```python
#Opening
opening = cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel)
plt.imshow(opening)
```

```python
#Closing
img = cv2.imread('/content/dog.jfif')[0:150][0:120]
plt.imshow(img)
closing = cv2.morphologyEx(img,cv2.MORPH_CLOSE,kernel)
plt.imshow(closing)
```
