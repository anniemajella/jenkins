```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
facecas=cv2.CascadeClassifier('/content/face.xml')
```

```python
image1 = cv2.imread('/content/zayn3.jpg')
image = cv2.cvtColor(image1,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
```

```python
face_img=image.copy()
```

```python
face_rect=facecas.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
```

```python
for (x,y,w,h) in face_rect:
  cv2.rectangle(face_img,(x,y),(x+w,y+h),(245,5,0),2)
```

```python
plt.imshow(face_img)
```
