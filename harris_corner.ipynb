```python
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("/content/chess.jpg")
```

```python
gary=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
dst=cv2.cornerHarris(gary,2,3,0,0.04)
```

```python
dst=cv2.dilate(dst,None)
```

```python
image[dst>0.01*dst.max()]=[255,0,0]
```

```python
plt.figure(figsize=(10,10))
```

```python
plt.imshow(image)
```
